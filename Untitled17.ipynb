```python
from audiosr import build_model, super_resolution
```

```python
audiosr = build_model(model_name="basic", device="auto")
```

```python
path = "footstep, running, snow-covered ground, boots, animated video game tiny girl character_4.wav"
```

```python
waveform = super_resolution(
    audiosr,
    path,
    seed=42,
    guidance_scale=3.5,
    ddim_steps=50,
    latent_t_per_second=12.8
)
```

```python
from audiosr import save_wave
```

```python
save_wave(waveform, "", name="output", samplerate=48000)
```
